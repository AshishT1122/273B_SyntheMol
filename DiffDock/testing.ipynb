```python
# Filtering the DataFrame by different 'Comment' types
active_df = df[df['Comment'] == 'active']
inactive_df = df[df['Comment'] == 'inactive']
inconclusive_df = df[df['Comment'] == 'inconclusive']

# Calculating Binding Affinity
active_df['Binding Affinity'] = active_df['Standard Value'].apply(lambda x: -np.log10(x) + 9)
inactive_df['Binding Affinity'] = inactive_df['Standard Value'].apply(lambda x: -np.log10(x) + 9)
inconclusive_df['Binding Affinity'] = inconclusive_df['Standard Value'].apply(lambda x: -np.log10(x) + 9)

plt.figure(figsize=(15, 8))

# Active
plt.subplot(1, 3, 1)
plt.hist(active_df['Binding Affinity'], bins=30, edgecolor='k', alpha=0.7, weights=np.ones(len(active_df)) / len(active_df))
plt.title('Active')
plt.xlabel('Binding Affinity (pChEMBL Value)')
plt.ylabel('Proportional Frequency')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0%}'))
plt.grid(True)

# Inactive
plt.subplot(1, 3, 2)
plt.hist(inactive_df['Binding Affinity'], bins=30, color='coral', edgecolor='k', alpha=0.7, weights=np.ones(len(inactive_df)) / len(inactive_df))
plt.title('Inactive')
plt.xlabel('Binding Affinity (pChEMBL Value)')
plt.ylabel('Proportional Frequency')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0%}'))
plt.grid(True)

# Inconclusive
plt.subplot(1, 3, 3)
plt.hist(inconclusive_df['Binding Affinity'], color='lime', bins=30, edgecolor='k', alpha=0.7, weights=np.ones(len(inconclusive_df)) / len(inconclusive_df))
plt.title('Inconclusive')
plt.xlabel('Binding Affinity (pChEMBL Value)')
plt.ylabel('Proportional Frequency')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0%}'))
plt.grid(True)

plt.tight_layout()
plt.show()
```
